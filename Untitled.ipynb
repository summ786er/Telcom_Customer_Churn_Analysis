```python
import matplotlib.pyplot as plt
import numpy as np

# ---- Data (no DataFrame) ----
senior = np.array([0, 1])
total = np.array([5901, 1142])
churned = np.array([1393, 476])
churn_pct = np.array([23.606168, 41.681261])

# Create figure with two y-axes
fig, ax1 = plt.subplots(figsize=(12, 8))

# X positions
x = np.arange(len(senior))
width = 0.6

# Bars
bars_total = ax1.bar(x - width/3, total, width/3,
                     label='Total Customers', alpha=0.6)
bars_churned = ax1.bar(x, churned, width/3,
                       label='Churned Customers', alpha=0.8)
bars_nonchurned = ax1.bar(x + width/3, total - churned, width/3,
                          label='Non-Churned Customers', alpha=0.8)

# Axis labels
ax1.set_xlabel('Senior Citizen Status', fontsize=12, fontweight='bold')
ax1.set_ylabel('Number of Customers', fontsize=12, fontweight='bold')
ax1.set_xticks(x)
ax1.set_xticklabels(['Non-Senior Citizen\n(0)', 'Senior Citizen\n(1)'], fontsize=11)

# Secondary axis for churn %
ax2 = ax1.twinx()
ax2.plot(x, churn_pct, marker='o', markersize=10, linewidth=3, label='Churn Rate (%)')
ax2.set_ylabel('Churn Rate (%)', fontsize=12, fontweight='bold')
ax2.set_ylim(0, 50)

# Value labels on bars
for bars in [bars_total, bars_churned, bars_nonchurned]:
    for bar in bars:
        h = bar.get_height()
        ax1.text(bar.get_x() + bar.get_width()/2, h + 20,
                 f'{int(h):,}', ha='center', fontsize=9, fontweight='bold')

# % labels
for xi, pct in zip(x, churn_pct):
    ax2.text(xi, pct + 1.5, f'{pct:.1f}%', ha='center', fontsize=11, fontweight='bold',
             bbox=dict(boxstyle='round,pad=0.2', facecolor='white', alpha=0.8))

# Title + grid
plt.title('Senior Citizen Churn Analysis: Counts vs Percentage',
          fontsize=14, fontweight='bold', pad=20)
ax1.grid(axis='y', alpha=0.3, linestyle='--')

# Legend
h1, l1 = ax1.get_legend_handles_labels()
h2, l2 = ax2.get_legend_handles_labels()
ax1.legend(h1 + h2, l1 + l2, loc='upper right', fontsize=10)

# Summary box
summary = (
    f'Key Insights:\n'
    f'• Senior Citizens: {churn_pct[1]:.1f}% churn rate\n'
    f'• Non-Senior: {churn_pct[0]:.1f}% churn rate\n'
    f'• Seniors are {churn_pct[1]/churn_pct[0]:.1f}x more likely to churn'
)
plt.text(1.02, 0.5, summary, transform=ax1.transAxes,
         fontsize=10, va='center',
         bbox=dict(boxstyle='round', facecolor='lightyellow', alpha=0.8))

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Create ChargesStatus groups
df["ChargesStatus"] = pd.qcut(
    df["TotalCharges"],
    q=3,
    labels=["Cheap", "Average", "Expensive"]
)

# Step 2: Total customers in each group
total_customers = df.groupby("ChargesStatus")["customerID"].count()

# Step 3: Total churn customers in each group
churn_customers = df[df["Churn"] == "Yes"].groupby("ChargesStatus")["customerID"].count()

# Step 4: Calculate churn percentage
churn_percentage = (churn_customers / total_customers) * 100

# Step 5: Create summary table
summary = pd.DataFrame({
    "Total Customers": total_customers,
    "Churn Customers": churn_customers,
    "Churn Percentage (%)": churn_percentage.round(2)
})

print("\nChurn Analysis Summary based on Charges Status:\n")
print(summary)

# Step 6: Bar Plot for churn percentage
plt.figure(figsize=(8,5))

plt.bar(
    summary.index,
    summary["Churn Percentage (%)"]
)

plt.title("Churn Percentage by Charges Status", fontsize=14)
plt.xlabel("Charges Status")
plt.ylabel("Churn Percentage (%)")

# Show values on bars
for i, v in enumerate(summary["Churn Percentage (%)"]):
    plt.text(i, v + 0.5, f"{v:.2f}%", ha='center', fontsize=11)

plt.show()

# Step 7: Business Insight
highest_churn_group = summary["Churn Percentage (%)"].idxmax()

print("\nBusiness Insight:")
print(f"• Customers in '{highest_churn_group}' category have the highest churn rate.")
print("• This group requires immediate retention strategies.")
print("• High churn indicates possible dissatisfaction, pricing issues, or competition.")
print("• Companies should focus on improving service quality and customer engagement in this segment.")
```
